```python
import nltk
nltk.download('brown')
from nltk.corpus import brown
```

```python
# Size OF Brown Corpus
print("Size of Brown Corpus",len(brown.words()))
```

```python
# Tokens in Brown Corpus
tokens = brown.words()
print("Tokens are: \n",tokens[1:10])
print("Number of tokens: ", len(tokens))
```

```python
# Categories in Brown Corpus
print("Categories:",brown.categories())
```

```python
#Size of word tokens
print("Number of word tokens: ",len(brown.words()))
```

```python
#Size of word types
print("Number of word types: ",len(set(brown.words())))
```

```python
# Size of the category “government”
print("The Size of the category government:",len(brown.words(categories='government')))
```

```python
# List the most frequent tokens
from nltk import FreqDist
fdist = FreqDist(brown.words())
print("The five most frequent tokens:",fdist.most_common(5))
```

```python
# The number of sentences
print("The number of sentences:",len(brown.sents()))
```

3. Create a text corpus with a minimum of 200 words (unique content). Implement the following text processing
• Word segmentation

• Sentence segmentation

• Convert to Lowercase

• Stop words removal

• Stemming

• Lemmatization

• Part of speech tagger

```python
text_corpus = "The quick brown fox jumps over the lazy dog. The dog is sleeping in the sun. The fox is hunting for food. The sun is shining brightly. The lazy dog is still sleeping. The quick fox is now eating a rabbit. The rabbit was caught in a trap. The fox is now full and satisfied. The lazy dog wakes up and sees the fox eating the rabbit. The dog is now jealous and starts to chase the fox."

# Word segmentation
from nltk.tokenize import word_tokenize

words = word_tokenize(text_corpus)
print(words)
```

```python
# Sentence segmentation
from nltk.tokenize import sent_tokenize

sentences = sent_tokenize(text_corpus)
print(sentences)
```

```python
# Convert to lowercase
text_corpus = text_corpus.lower()
print(text_corpus)
```

```python
# Stop words removal
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))
filtered_words = [word for word in word_tokenize(text_corpus) if word not in stop_words]
print(filtered_words)
```

```python
# Stemming
from nltk.stem import PorterStemmer
```

```python
stemmer = PorterStemmer()
stemmed_words = [stemmer.stem(word) for word in filtered_words]
print(stemmed_words)
```

```python
# Lemmatization
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
lemmatizer = WordNetLemmatizer()
lemmatized_words = [lemmatizer.lemmatize(word) for word in filtered_words]
print(lemmatized_words)
```

```python
# Part of speech tagger
from nltk import pos_tag
nltk.download('averaged_perceptron_tagger')
```

```python
pos_tagged_words = pos_tag(filtered_words)
print(pos_tagged_words)
```
